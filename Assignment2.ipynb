# Assignment 2

Create a program  that can save contacts also able to show saved contacts. Each contact only consisting name and its phone number. Once it's running, a system would give a greeting at first following with the menu. Program will be automatically back to the menu after adding and/or view contact list until it requested to exit.

```python
def show_menu():
    print('''
    -------Menu-------
    1. Contact list
    2. Add to contact
    3. Exit
    ''')

contact_name = []
contact_number = []

# Introduction
print("Welcome!")

# Showing the chose menu
while True:
    show_menu()
    chosen_menu = input('Please choose the menu: ')
    
    if chosen_menu == '1':
        print('\nContact List: ')
        for x in range(len(contact_name)):
            print(' ')
            print('Name: {}'.format(contact_name[x]))
            print('Phone Number: {}'.format(contact_number[x]))
    elif chosen_menu == '2':
        name = input('\nName: ')
        number = input('Phone Number: ')
        print('\nContact successfully added to contact list!')
        contact_name.append(name)
        contact_number.append(number)
    elif chosen_menu == '3':
        print('\nThe program is over, cheerio!')
        print('===============================')
        break
    else:
        print('Sorry, the chosen menu is unavailable')
```

```python
def show_menu2():
    print('Please input a menu that i can help: ')
    print('1. Calculator')
    print('2. Confession')
    print('3. Exit')
    
# Introduction
print('Hello, I am AIs. I am your first virtual assistant')
print('What is your name?')

name2 = input('Input your name: ')

print("\nHello {}!".format(name2))
print('Is there anything i can help?')

# Showing chosen menu
while True:
    print(' ')
    show_menu2()
    chosen_menu2 = input('Please choose the menu: ')
    
    if chosen_menu2 == '1':
        while True:
            operation = input('''
            Please type in the math operation you would like to complete:
            + for addition
            - for subtraction
            * for multiplication
            / for division
            x for exit
            ''')
            
            if operation == '+':
                # Addition
                number1 = int(input('\nFirst number: '))
                number2 = int(input('Second number: '))
                print('\n{} + {} = '.format(number1, number2), number1+number2)
    
            elif operation == '-':
                # Subtraction
                number1 = int(input('\nFirst number: '))
                number2 = int(input('Second number: '))
                print('\n{} - {} = '.format(number1, number2), number1-number2)
    
            elif operation == '*':
                # Multiplication
                number1 = int(input('\nFirst number: '))
                number2 = int(input('Second number: '))
                print('\n{} * {} = '.format(number1, number2), number1*number2)
    
            elif operation == '/':
                # Division
                number1 = int(input('\nFirst number: '))
                number2 = int(input('Second number: '))
                print('\n{} / {} = '.format(number1, number2),  number1/number2)
    
            elif operation == 'x':
                # Exit
                print("\n Thank you")
                break
            else:
                print("\n Sorry, your input is invalid")
    
    elif chosen_menu2 == '2':
        input('\nTell me what is matter? ')
        while True:
                story = input('Please tell me more: ')
                
                if 'sad' in story or 'failed' in story or 'losed' in story or 'terrible' in story:
                    print('Why did you think so? ')
                elif 'dissapointed' in story or 'ashamed' in story or 'hopeless' in story:
                    print('What can cause you like this? ')
                elif 'happy' in story or 'delighted' in story or 'awesome' in story:
                    print('That is amazing, glad to hear that!')
                else:
                    print('Alright! everything will be definitely okay, dearest', ':D')
                    break
    elif chosen_menu2 == '3':
        print('\nOkay, see you later!')
        print('====================')
        break
    else:
        print('\nSorry, i do not understand what you want me to')
```
