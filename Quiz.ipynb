```python
import datetime

agenda_list = []
agenda_date = []
agenda_time = []

def show_menu():
    print('\nPlease input a menu that i can help (numbers only): ')
    print('1. About today')
    print("2. Record today's agenda")
    print('3. Quit')

def show_abouttoday():
    print('Please input a menu that i can help (numbers only): ')
    print('1. What day is it now?')
    print('2. What date is it now?')
    print('3. Previous menu')
    print('4. Quit')
    
def show_agenda():
    print('Please input a menu that i can help (numbers only): ')
    print('1. Agenda list')
    print('2. Add new agenda')
    print('3. Previous menu')
    print('4. Quit')

def empty_agenda():
    if not agenda_list:
        print("=================================")
        print("Your agenda list is empty!\nPlease add new agenda")
        print("=================================")

def add_agenda():
    agenda = input("What is your plan? ")
    date = input("What date? ")
    time = input("What time? ")
    agenda_list.append(agenda)
    agenda_date.append(date)
    agenda_time.append(time)
    print("\n===========================")
    print("Agenda is successfully added")
    print("===========================")
    
def show_agendalist():
    print("\n Agenda list: ")
    for i in range(len(agenda_list)):
        print("-----------------------------")
        print("Agenda: {}".format(agenda_list[i]))
        print("Will be held on: {}".format(agenda_date[i]))
        print("At time: {}".format(agenda_time[i]))
        print("-----------------------------")

def create_agendalist():
    f = open("Agenda List.txt", "w")
    for x in range(len(agenda_list)):
        f.write("Agenda: {} \nWill be held on: {}\nAt time: {}\n".format(agenda_list[x], agenda_date[x], agenda_time[x]))
    f.close()

print("Hello, I am AIs, your first virtual assistant\nWhat is your name?")
name = input("Insert your name: ")

print("\nHello, {}!".format(name))
print('Is there anything i can help?')

while True:
    
    show_menu()
    menu = input("\nPlease choose a menu: ")
    
    if menu == "1":
        show_abouttoday()
        about = input("\nPlease choose a menu: ")
        if about == "1":
            x = datetime.datetime.now()
            day = x.strftime("%A")
            print("==================================")
            print("Seriously? You must've known that! Today is "+ day )
            print("==================================")
        elif about == "2":
            print("==================================")
            print("Today is " + x.strftime("%d-%m-%Y"))
            print("==================================")
        elif about == "3":
            print(" ")
        elif about == "4":
            print("==================================")
            print("Alright, see you soon!")
            print("==================================")
            break
        else:
            print("Sorry, i do not understand what you want me to")
    elif menu == "2":
        show_agenda()
        agenda = input("\nPlease choose a menu: ")
        if agenda == "1":
            show_agendalist()
            empty_agenda()
        elif agenda == "2":
            add_agenda()
        elif agenda == "3":
            print(" ")
        elif agenda == "4":
            print("================================")
            print("Alright, see you soon!")
            print("================================")
            break
        else:
            print("Sorry, i do not understand what you want me to")
    elif menu == "3":
        print("Alright, see you soon!")
        create_agendalist()
        break
    else:
        print("Sorry, menu doesn't exist")
```
